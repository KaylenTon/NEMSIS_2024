```python
import os
import pandas as pd

def load_sas_file(sas_file, output_dir_name, chunk_size = 10000):
    """Loads large SAS files in chunks.
    PARAMETERS:
    sas_file = file ending with .sas7bdat
    output_dir_name = the name of the folder directory containing the parquet files
    chunk_size = takes the # of rows based on chunk size at a time. Large chunk size loads files faster, but you'll need more RAM. For limited RAM, use smaller chunk size. Default is 10000"""
    file_path = f'{sas_file}'

    sas_chunks = pd.read_sas(file_path, chunksize = chunk_size)
    os.makedirs(output_dir_name, exist_ok=True)

    for i, chunk in enumerate(sas_chunks):
        chunk_path = os.path.join(output_dir_name, f'chunk_{i}.parquet')
        chunk.to_parquet(chunk_path, index = False)
    
    return()

#load_sas_file("computedelements.sas7bdat", 'computed_elements') <- sample function use
```
